## Exercise 1
Jane paid 12 dollars for 4 cups of coffee and 4 cups of tea. 3 cups of coffee cost as much as 2 cups of tea. What would be the total cost of 5 cups of coffee and 5 cups of tea?

### Solution

> Let $x$ be the unit price of coffee and $y$ be the unit price of tea

```python
# Your solution here 
4x + 4y = 12
3x = 2y
x = 2y/3

# To find the total cost of 5 cups of coffee and 5 cups of tea, we need to substitute x = 2y/3 into the first equation:

4(2y/3) + 4y = 12
8y/3 + 4y = 12
(8y + 12y)/3 = 12
20y/3 = 12

# Now we can solve for y:
20y = 36
y = 36/20
y = 1.8

# Now substitute the value of y back into the first equation to solve for x:
4x + 4(1.8) = 12
4x + 7.2 = 12
4x = 4.8
x = 4.8/4
x = 1.2

# So the unit price of coffee is 1.2 and the unit price of tea is 1.8. 

# To find the total cost of 5 cups of coffee and 5 cups of tea:
5(1.2) + 5(1.8) = 6 + 9 = 15

# Therefore, the total cost of 5 cups of coffee and 5 cups of tea is 15 dollars.



# Answer: 5 cups of tea and 5 cups of coffee = 15 dollars
```

## Exercise 2

Jim has more money than Bob. If Jim gave Bob 20 dollars, they would have the same amount. If Bob gave Jim 22 dollars, however, Jim would then have twice as much as Bob. 

How much does each one actually have?

### Solution
> Let x be the amount of money that Jim has and y be the amount that Bob has 

```python
# Your solution here 
# Let's set up the equations based on the given information:

# If Jim gave Bob 20 dollars, they would have the same amount
x - 20 = y + 20

# If Bob gave Jim 22 dollars, however, Jim would then have twice as much as Bob
x + 22 = 2 * (y - 22)

# Now, let's solve this system of equations to find the values of x and y

# From the first equation, we can rewrite it as x - y = 40
# From the second equation, we can rewrite it as x - 2y = -66

x - y = 40
x - 2y = -66

# We can solve this system of equations using elimination method.
# Multiply the first equation by 2 to eliminate x:
2*(x - y) = 2*40
2x - 2y = 80

# Now we can subtract the second equation from this:
2x - 2y - (x - 2y) = 80 - (-66)
x = 146

# Substitute x = 146 back into the first equation to solve for y:
146 - y = 40
y = 146 - 40
y = 106

# Therefore, Bob has 106 dollars and Jim has 146 dollars.

# Answer:
# y = 106 (Bob's amount)
# x = 146 (Jim's amount)
```

## Exercise 3

Mia has 30 coins, consisting of quarters (25 cents) and dimes (10 cents), which totals to the amount 5.70 dollars.  
How many of each does she have?

### Solution

> Let x be the number of quarters and y be the number of dimes 

```python
# Your solution here 
# Your solution here 
# Let's set up the equations based on the given information:

# Let x be the number of quarters and y be the number of dimes 
# The total value of quarters in cents is 25x
# The total value of dimes in cents is 10y
# The total number of coins is x + y

# The total value of coins in cents is 25x + 10y
# The total value of coins in dollars is 0.01(25x + 10y)

# The total value of the coins is 5.70 dollars, so we can write the equation:
0.01(25x + 10y) = 5.70

# Simplifying the equation:
0.25x + 0.1y = 5.70

# Now, we know that Mia has 30 coins, so we can write another equation:
x + y = 30

# Now, we have a system of linear equations:
0.25x + 0.1y = 5.70
x + y = 30

# Let's solve this system of equations to find the values of x and y

# We can multiply the second equation by 0.1 to eliminate y:
0.1(x + y) = 0.1(30)
0.1x + 0.1y = 3

# Now, we can subtract this equation from the first equation:
0.25x + 0.1y - (0.1x + 0.1y) = 5.70 - 3
0.15x = 2.70
x = 2.70/0.15
x = 18

# Substitute x = 18 back into the second equation to solve for y:
18 + y = 30
y = 30 - 18
y = 12

# Therefore, Mia has 18 quarters and 12 dimes.

# Answer:
# x = 18 quarters
# y = 12 dimes
```
